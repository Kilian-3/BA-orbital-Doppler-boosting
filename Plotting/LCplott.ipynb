## Daten holen mit numpy

```python
paths = ['1_4fgl_j0457.0-2324_lightcurve.npy', '1_4fgl_j0721.9+7120_lightcurve.npy', 
         '1_4fgl_j0811.4+0146_lightcurve.npy', '1_4fgl_j1555.7+1111_lightcurve.npy', 
         '1_4fgl_j2158.8-3013_lightcurve.npy']
```

```python
tmin = []
flux = []
flux_err = []

for p in paths:
    n = np.load(standardpath + p, allow_pickle=True,encoding='latin1').flat[0]
    tmin.append(n['tmin'])
    flux.append(n['flux'])
    flux_err.append(n['flux_err'])
```

### time transformation to MJD from MET

```python
t = TimeDelta(tmin,format='sec')
t.size
start_time = Time('2001-01-01 00:00:00',format='iso')
time = start_time + t
time = time.mjd
```

```python
len(time[0])
```

## plotting

```python
fig = plt.figure(figsize=(8.27, 11)) # DinA 4: 8.27*1.5, 11.69*1.5
fig.tight_layout()

gs = mpl.gridspec.GridSpec(5, 4)
ax0 = fig.add_subplot(gs[0, :])
plot0 = ax0.errorbar(time[0], flux[0]*1e8,yerr=flux_err[0]*1e8,xerr=np.linspace(14,14,len(time[0])),
                     marker='.', color='darkblue', markersize = 2,linestyle='',capsize=1,
                     elinewidth=0.5) 

ax1 = fig.add_subplot(gs[1, :], sharex = ax0) 
plot1 = ax1.errorbar(time[1], flux[1]*1e8,yerr=flux_err[1]*1e8,xerr=np.linspace(14,14,len(time[1])),
                     marker='.', color='darkblue', markersize = 2,linestyle='',capsize=1,
                     elinewidth=0.5) 

ax2 = fig.add_subplot(gs[2, :], sharex = ax0) 
plot2 = ax2.errorbar(time[2], flux[2]*1e8,yerr=flux_err[2]*1e8,xerr=np.linspace(14,14,len(time[2])),
                     marker='.', color='darkblue', markersize = 2,linestyle='',capsize=1,
                     elinewidth=0.5) 

ax3 = fig.add_subplot(gs[3, :], sharex = ax0) 
plot3 = ax3.errorbar(time[3], flux[3]*1e8,yerr=flux_err[3]*1e8,xerr=np.linspace(14,14,len(time[3])),
                     marker='.', color='darkblue', markersize = 2,linestyle='',capsize=1,
                     elinewidth=0.5) 

ax4 = fig.add_subplot(gs[4, :], sharex = ax0) 
plot4 = ax4.errorbar(time[4], flux[4]*1e8,yerr=flux_err[4]*1e8,xerr=np.linspace(14,14,len(time[4])),
                     marker='.', color='darkblue', markersize = 2,linestyle='',capsize=1,
                     elinewidth=0.5) 

# ax0.set_xlim(54000,60500) 
# ax1.set_xlim(54000,60500) 
# ax2.set_xlim(54000,60500) 
# ax3.set_xlim(54000,60500) 
# ax4.set_xlim(54000,60500) 

ax0.set_ylabel(r'Flux $[\mathrm{ph}/\mathrm{cm}^2/\mathrm{s}]\cdot10^{-8}$') 
ax1.set_ylabel(r'Flux $[\mathrm{ph}/\mathrm{cm}^2/\mathrm{s}]\cdot10^{-8}$')
ax2.set_ylabel(r'Flux $[\mathrm{ph}/\mathrm{cm}^2/\mathrm{s}]\cdot10^{-8}$')
ax3.set_ylabel(r'Flux $[\mathrm{ph}/\mathrm{cm}^2/\mathrm{s}]\cdot10^{-8}$')
ax4.set_ylabel(r'Flux $[\mathrm{ph}/\mathrm{cm}^2/\mathrm{s}]\cdot10^{-8}$')

ax4.set_xlabel('Time [MJD]') 

plt.setp(ax0.get_xticklabels(), visible=False) 
plt.setp(ax1.get_xticklabels(), visible=False) 
plt.setp(ax2.get_xticklabels(), visible=False) 
plt.setp(ax3.get_xticklabels(), visible=False) 

plt.subplots_adjust(hspace=.0) 
plt.locator_params(axis='x', nbins=7) 

ax0.set_ylim(ax0.get_ylim()[0],ax0.get_ylim()[1]*1.2) 
ax1.set_ylim(ax1.get_ylim()[0],ax1.get_ylim()[1]*1.2) 
ax2.set_ylim(ax2.get_ylim()[0],ax2.get_ylim()[1]*1.2) 
ax3.set_ylim(ax3.get_ylim()[0],ax3.get_ylim()[1]*1.2) 
ax4.set_ylim(ax4.get_ylim()[0],ax4.get_ylim()[1]*1.2) 

ax0.tick_params(axis="x", direction="inout") 
ax1.tick_params(axis="x", direction="inout") 
ax2.tick_params(axis="x", direction="inout") 
ax3.tick_params(axis="x", direction="inout") 
ax4.tick_params(axis="x", direction="inout") 

ax0.tick_params(axis="y", direction="inout") 
ax1.tick_params(axis="y", direction="inout") 
ax2.tick_params(axis="y", direction="inout") 
ax3.tick_params(axis="y", direction="inout") 
ax4.tick_params(axis="y", direction="inout") 

ax0.grid(color='darkgrey', linestyle='-', linewidth=0.25) 
ax1.grid(color='darkgrey', linestyle='-', linewidth=0.25) 
ax2.grid(color='darkgrey', linestyle='-', linewidth=0.25) 
ax3.grid(color='darkgrey', linestyle='-', linewidth=0.25) 
ax4.grid(color='darkgrey', linestyle='-', linewidth=0.25) 

fig.align_labels() 

ax0.legend([plot0],['4FGL J0457.0-2324, 28-day binning'], loc='upper right',fontsize=10) 
ax1.legend([plot1],['4FGL J0721.9+7120, 28-day binning'], loc='upper right',fontsize=10) 
ax2.legend([plot2],['4FGL J0811.4+0146, 28-day binning'], loc='upper right',fontsize=10) 
ax3.legend([plot3],['4FGL J1555.7+1111, 28-day binning'], loc='upper right',fontsize=10) 
ax4.legend([plot4],['4FGL J2158.8-3013, 28-day binning'], loc='upper right',fontsize=10) 

#plt.savefig(standardpath+'multiLC.png',dpi=300) 
plt.show() 
```
